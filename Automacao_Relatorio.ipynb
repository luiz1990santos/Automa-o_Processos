```python
vendas = pd.read_excel('Vendas - Dez.xlsx')

display(vendas)
```

```python
vendas.count()
```

```python
faturamento = vendas['Valor Final'].sum()

print(faturamento)

produtos = vendas['Quantidade'].sum()

print(f'{produtos}')

media = vendas['Valor Final'].mean()

print(f'{media:,.3f}')

maior_vendas = vendas['Produto'].max()

print(maior_vendas)

menos_vendas = vendas['Produto'].min()

print(menos_vendas)

shopping_mais = vendas['ID Loja'].max()

print(shopping_mais)

shopping_menos = vendas['ID Loja'].min()

print(shopping_menos)
```
